<h1>Limpieza de datos</h1>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
#Cargar csv de personal population

personal_Victimization_mapped = pd.read_csv("../../data/personal_victimization_mapped.csv")
```

```python
personal_Victimization_mapped
```

```python
#SEPARAR LAS X DE LA Y
y = personal_Victimization_mapped["newoff"]
x = personal_Victimization_mapped.drop(columns=["newoff"])
```

```python
x.columns
```

```python
y.value_counts()
```

Tenemos 5 posibles valores en Y, y como observamos la cantidad de datos no está proporcionada, por esto a la hora de dividir en Train y Test vamos a usar 'stratify = y'

<h3> Separar los  datos en Train y Test </h3>

Se utilizó train test split con una semilla de 42, dejando la misma proporcion de 'y' en test y train y se dividio el conjunto en un 80 para train y 20 para test

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y) 
```

```python
print(f"\n Datos de entrenamiento: {len(x_train)} crimenes ({len(x_train)/len(personal_Victimization_mapped)*100:.1f}%)")
print(f" Datos de prueba: {len(x_test)} crimenes ({len(x_test)/len(personal_Victimization_mapped)*100:.1f}%)")
```

<h3>Enconding de categóricas</h3>

Vamos a realizar one hot encoding con las variables categóricas nominales, osea las que no tienen un orden logico.

```python
var_categoricas = ['ager', 'sex', 'race_ethnicity', 'hincome1', 'marital', 'popsize','region', 'msa', 'educatn1', 'locationr']
```

```python
x_train = pd.get_dummies(x_train, columns= var_categoricas,drop_first=True)
```

```python
x_test = pd.get_dummies(x_test, columns= var_categoricas,drop_first=True)
```

Mapear el target

```python
# TYPE OF CRIME 
y_train= y_train.map({ 
    'Rape/sexual assault' : 1, 
    'Robbery' : 2, 
    'Aggravated assault' : 3, 
    'Simple assault' : 4, 
    'Personal theft/larceny' :5
})
```

```python
# TYPE OF CRIME 
y_test = y_test.map({ 
    'Rape/sexual assault' : 1, 
    'Robbery' : 2, 
    'Aggravated assault' : 3, 
    'Simple assault' : 4, 
    'Personal theft/larceny' :5
})
```

```python
# Guardar x_train, x_test, y_train, y_test en CSV sin escalar para el arbol de desicion y el random forest
x_train.to_csv("../../data/x_train.csv", index=False)
x_test.to_csv("../../data/x_test.csv", index=False)
y_train.to_csv("../../data/y_train.csv", index=False)
y_test.to_csv("../../data/y_test.csv", index=False)
```
